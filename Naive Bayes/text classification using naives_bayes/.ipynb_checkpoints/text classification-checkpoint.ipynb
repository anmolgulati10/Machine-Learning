```python
#for opening the file and saving it in array
import numpy as np
import os
import re
x=[]
y=[]
for category in os.listdir("C:/Users/shivani/Downloads/20_newsgroups"):
    for document in os.listdir(f"C:/Users/shivani/Downloads/20_newsgroups/{category}"):
            with open(f"C:/Users/shivani/Downloads/20_newsgroups/{category}/{document}",'r') as f:
                      x.append(f.read())
                      y.append(category)

```

```python
#for removing all the symbols and converting the data into the uppercase

for i in range(len(x)):
    x[i]=re.sub("[^a-zA-Z]"," ",x[i])
    x[i]=x[i].upper()
    x[i]=x[i].split(" ")
    
x=np.array(x)
y=np.array(y)
x
```

```python
#splitting the data into training and testing part
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#for removing the '' value from words from both training and testing data
for i in range(len(x_train)):
    k=x_train[i].count('')
    for j in range(k):
        x_train[i].remove('')
for i in range(len(x_test)):
    k=x_test[i].count('')
    for j in range(k):
        x_test[i].remove('')
```

```python
#for making dictionary for occurence of words from training data
dictionary={}
for i in range(len(x_train)):
    for j in x_train[i]:
        if j in dictionary:
            dictionary[j]=dictionary[j]+1
        else:
            dictionary[j]=1
        
```

```python
#for sorting the dictionary by value and again converting into dictionary as sorted function returns tuple of length 2
import operator
dictionary=sorted(dictionary.items(),key=operator.itemgetter(1),reverse=True)
dictionary=dict(dictionary)
dictionary
```

```python
#for opening the file which contain general stop words of english language
import re
import os
with open("C:/Users/shivani/Desktop/data science/text classification using naives_bayes/data.txt",'r') as f:
    z=f.read()
z=re.sub("[^a-zA-Z]"," ",z)
z=z.upper()
z=z.split(" ")
z
```

```python
#removing the stop words from the old dictionary using pop function
for i in z:
    if i in dictionary:
        dictionary.pop(i)
print(len(dictionary))
dictionary
```

```python
#slicing the sorted dictionary for first 1000 keys
f_dict={}
i=0
for k,v in dictionary.items():
    i=i+1
    f_dict[k]=v
    if i==1000:
        break
print(len(f_dict))
f_dict
```

```python
#decoding data as to convert into frequency of occurence of word in both training and testing data
def creating_data(x,f_dict):
    y=[]
    for i in range(len(x)):
        y.append([])
        for j in f_dict:
                c=x[i].count(j)
                y[i].append(c)
    return y
```

```python
x_train=creating_data(x_train,f_dict)
x_test=creating_data(x_test,f_dict)
```

```python
#converting the list into np.array 
x_train=np.array(x_train)
x_test=np.array(x_test)
```

```python
#custom fit function for the naive bayes theorem
def fit(x_train,y_train,f_dict):
    result={}
    all_classes=set(y_train)
    result['total']=len(y_train)
    for a in all_classes:
        result[a]={}
        z=(y_train==a)
        x_train_p=x_train[z]
        y_train_p=y_train[z]
        result[a]["total"]=len(y_train_p)
        i=0
        for k in f_dict.keys():
            result[a][k]=x_train_p[:,i].sum()
            i=i+1
    return result
                
        
```

```python
#custom predict function for the testing data of naive bayes theorm 
def predict(x_test,result):
    y_pred=[]
    for x in x_test:
        y=predictsinglepoint(x,result)
        y_pred.append(y)
    return y_pred
```

```python
#function for choosing best class
def predictsinglepoint(x,result):
    max_prob=-1000000
    feature=""
    for c in result.keys():
        if c=="total":
            continue
        else:
            prob=probability(x,result,c)
            if prob>=max_prob:
                max_prob=prob
                feature=c
    return feature
```

```python
#function to calculate probability of class
def probability(x,result,c):
    output=np.log(result[c]["total"])-np.log(result["total"])
    den=0
    for k in result[c].keys():
        if k=="total":
            continue
        else:
            den=den+result[c][k] +1
    i=0
    for k in result[c].keys():
        if k=="total":
            continue
        elif x[i]==0:
            i=i+1
            continue
        else:
            
            num=result[c][k]+1
            p=np.log(x[i])+np.log(num)-np.log(den)
            output=output+p
            i=i+1
    return output
```

```python
result=fit(x_train,y_train,f_dict)
y_pred=predict(x_test,result)
```

```python
#classification report for custom naive bayes algorithm
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
