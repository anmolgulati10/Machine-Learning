```python
from sklearn import datasets
import pandas as pd
iris=datasets.load_iris()
```

```python
x=pd.DataFrame(iris.data)
y=iris.target
y.shape
```

```python
def label(x,j):
    second=x[j].mean()
    min=x[j].min()
    max=x[j].max()
    first=(min+second)/2
    third=(second+max)/2
    return x[j].apply(f,args=(first,second,third))

    
    
```

```python
def f(val,*b):
    if val<b[0]:
        return 'a'
    elif val<b[1]:
        return 'b'
    elif val<b[2]:
        return 'c'
    else:
        return 'd'
```

```python
for j in range(x.shape[1]):
    x[j+4]=label(x,j)
```

```python
x.drop([0,1,2,3],inplace=True,axis=1)
x
```

```python
import numpy as np
x=np.array(x)
y=np.array(y)
y
```

```python
y.shape
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
def fit(x_train,y_train):
    result={}
    classes=set(y_train)
    result["total"]=len(y_train)
    for a in classes:
        result[a]={}
        features=x_train.shape[1]
        result[a]["total"]=(y_train==a).sum()
        for j in range(features):
            result[a][j]={}
            possible_classes_column=set(x_train[:,j])
            for fl in possible_classes_column:
                result[a][j][fl]=(x_train[(y_train)==a][:,j]==fl).sum()
    return result   
```

```python
def predict(dictionary,x_test):
    y_pred=[]
    for x in x_test:
        best_class=predictsinglepoint(dictionary,x)
        y_pred.append(best_class)
    return y_pred
            
```

```python
def predictsinglepoint(dictionary,x):
    possible_classes=dictionary.keys()
    best_class=""
    max_probability=-100
    for p in possible_classes:
        if p=="total":
            continue
        else:
            prob=probability(dictionary,x,p)
            if prob>=max_probability:
                max_probability=prob
                best_class=p
    return best_class
```

```python
def probability(dictionary, x, current_class):
    output = np.log(dictionary[current_class]["total"]) - np.log(dictionary["total"])
    num_features = len(dictionary[current_class].keys()) - 1;
    for j in range( num_features ):
        xj = x[j - 1]
        count_current_class_with_value_xj = dictionary[current_class][j][xj] + 1
        count_current_class = dictionary[current_class]["total"] + len(dictionary[current_class][j].keys())
        current_xj_probablity = np.log(count_current_class_with_value_xj) - np.log(count_current_class)
        output = output + current_xj_probablity
    return output
```

```python
from sklearn.metrics import confusion_matrix,classification_report 
dictionary=fit(x_train,y_train)
print(dictionary)
y_pred=predict(dictionary,x_test)
```

```python
print(classification_report(y_test,y_pred))
```
