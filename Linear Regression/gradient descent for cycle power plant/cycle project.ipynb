```python
import numpy as np
import pandas as pd
data=np.loadtxt("x_y_train.csv",delimiter=",")
x_test=np.loadtxt("x_test.csv",delimiter=",")
print(data.shape)
print(x_test.shape)
```

```python
x_train=data[:,0:(data.shape)[1]-1]
y_train=data[:,(data.shape)[1]-1]
print(x_train.shape)
print(y_train.shape)
```

```python
x=np.ones(len(data)).reshape(-1,1)
x_train=np.concatenate((x_train,x),axis=1)
print(x_train.shape)
```

```python
data=np.concatenate((x_train,y_train.reshape(-1,1)),axis=1)
data.shape
```

```python
def run(data,x_test):
    num_iteration=5000
    learning_rate=.00000093
    m=gd(data,num_iteration,learning_rate)
    y_pred=predict(x_test,m)
    np.savetxt("y_final1.csv",y_pred,fmt="%1.5f")
```

```python
def gd(data,num_iteration,learning_rate):
    m=np.zeros((data.shape)[1]-1)
    a=0
    for i in range(num_iteration):
        m=step_gradient(data,m,learning_rate)
        a=a+1
        print("cost:",cost(data,m),end=" ")
        print(a)
    return m 
```

```python
def step_gradient(data,m,learning_rate):
    new_m=np.zeros(len(m))
    N=len(data)
    for k in range(len(m)):
        sum=0
        for i in range(N):
            x=data[i,0:(data.shape)[1]-1]
            y=data[i,(data.shape)[1]-1]
            sum+=(-2/N)*(y-((m*x).sum()))*x[k]
        new_m[k]=m[k]-(learning_rate*sum)
    return new_m
```

```python
def predict(x_test,m):
    y_pred=np.zeros(len(x_test))
    for i in range(len(x_test)):
        x=x_test[i,0:(x_test.shape)[1]]
        y_pred[i]=(m*x).sum()
    return y_pred
```

```python
def cost(data,m):
    sum=0
    N=len(data)
    for i in range(len(data)):
        x=data[i,0:(data.shape[1])-1]
        y=data[i,(data.shape)[1]-1]
        sum+=((y-((m*x).sum()))**2)
    return (1/N)*sum
        
        
```

```python
run(data,x_test)
```
