```python
def step_gradient(data,m,learning_rate):
    new_m=np.zeros(len(m))
    N=len(data)
    for k in range((data.shape[1])-1):
        sum=0
        for i in range(len(data)):
            x=data[i,0:(data.shape[1])-1]
            y=data[i,(data.shape[1])-1]
            sum+=(-2/N)*(y-((m*x).sum()))*x[k]
        new_m[k]=m[k]-(learning_rate*sum)
    return new_m
```

```python
def cost(data,m):
    N=len(data)
    sum=0
    for i in range(N):
        x=data[i,0:(data.shape[1])-1]
        y=data[i,(data.shape[1]-1)]
        sum+=(y-(m*x).sum())
    return (1/N)*sum
```
