```python
def run(data):
    num_iteration=1000
    learning_rate=0.1
    m=gd(data,num_iteration,learning_rate)
    print(*m)
```

```python
def gd(data,num_iteration,learning_rate):
    m=np.zeros((data.shape)[1]-1)
    for i in range(num_iteration):
        m=step_gradient(data,m,learning_rate)
        cost(data,m)
        print("cost:",cost(data,m))
    return m
```

```python
def step_gradient(data,m,learning_rate):
    new_m=np.zeros(len(m))
    N=len(data)
    for k in range(len(m)):
        sum=0
        for i in range(N):
            x=data[i,0:(data.shape)[1]-1]
            y=data[i,(data.shape)[1]-1]
            z=math.exp((m*x).sum())
            sum+=(-1/N)*(y-(z/(1+z)))*x[k]
        new_m[k]=m[k]-(learning_rate)*sum
    return new_m
```

```python
def cost(data,m):
    import math
    sum=0
    N=len(data)
    for i in range(N):
        x=data[i,0:(data.shape)[1]-1]
        y=data[i,(data.shape)[1]-1]
        z=math.exp((m*x).sum())
        #print(z)
        sum+=(-1/N)*(y*(math.log((z/(z+1))))+((1-y)*(math.log((1-(z/(z+1)))))))
    return sum
```
